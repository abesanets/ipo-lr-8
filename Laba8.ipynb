### **Задание 3**, построение графика

Импортируем **pandas** которая нужна для работы с таблицами<br>
А также **matplotlib** для построения графиков

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Загружаем данные из Excel
df = pd.read_excel('students.xlsx')

# 2. Список предметов
subjects = ['Математика', 'Физика', 'Информатика']

# 3. Вычисляем процент учащихся с отметкой ниже 6
percent_below_6 = {}
for subject in subjects:
    total_students = len(df)
    below_6 = (df[subject] < 6).sum()
    percent_below_6[subject] = (below_6 / total_students) * 100

# 4. Преобразуем в DataFrame для удобства построения графика
percent_df = pd.DataFrame(list(percent_below_6.items()), columns=['Предмет', 'Процент <6'])

# 5. Строим столбчатую диаграмму
plt.figure(figsize=(5,5))
plt.bar(percent_df['Предмет'], percent_df['Процент <6'], color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Процент учащихся с отметкой ниже 6 по предметам')
plt.ylabel('Процент учащихся (%)')
plt.xlabel('Предмет')
plt.ylim(0, 100)
plt.show()
```

### **Задание 4**, запись в файл

```python
import pandas as pd

# 1. Загружаем данные из Excel
df = pd.read_excel('students.xlsx')

# 2. Список предметов
subjects = ['Математика', 'Физика', 'Информатика']

# 3. Вычисляем процент учащихся с отметкой ниже 6
percent_below_6 = {}
for subject in subjects:
    total_students = len(df)
    below_6 = (df[subject] < 6).sum()
    percent_below_6[subject] = (below_6 / total_students) * 100

# 4. Преобразуем в DataFrame
percent_df = pd.DataFrame(list(percent_below_6.items()), columns=['Предмет', 'Процент <6'])

# 5. Сохраняем в Excel
percent_df.to_excel('students_result.xlsx', index=False)
print("Результаты сохранены в 'students_result.xlsx'")
```
